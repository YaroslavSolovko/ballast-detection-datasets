## STEP 1. **Data Understanding & Profiling**

```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr, pearsonr
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif

df = pd.read_csv('Sparse_Features_Subset.csv')

print(f"Shape: {df.shape}")
print(df.head())

# Define targets

target_numeric = 'T5_2_TP'
target_categorical = 'UR_Category'

# Drop rows with missing target to ensure valid MI/correlation
df = df.dropna(subset=[target_numeric, target_categorical])

# Initialize results storage
results = []

# Loop through each sparse feature

# Exclude GUID and targets
features = df.columns.difference(['GUID', target_numeric, target_categorical])

for feature in features:
    col = df[feature]
    
    # % missing
    pct_missing = col.isnull().mean() * 100
    
    # % zeros (for sparsity confirmation, optional)
    pct_zero = (col == 0).mean() * 100 if np.issubdtype(col.dtype, np.number) else np.nan
    
    # % non-zero
    pct_non_zero = 100 - pct_missing - pct_zero if not np.isnan(pct_zero) else np.nan

    # unique count
    unique_count = col.nunique(dropna=True)
    
    # variance
    variance = col.var(ddof=0)
    
    # Pearson & Spearman correlation with numeric target
    if np.issubdtype(col.dtype, np.number):
        corr_pearson = pearsonr(col.fillna(0), df[target_numeric].fillna(0))[0]
        corr_spearman = spearmanr(col.fillna(0), df[target_numeric].fillna(0))[0]
    else:
        corr_pearson = np.nan
        corr_spearman = np.nan
    
    # Mutual Information with numeric target
    try:
        mi_numeric = mutual_info_regression(col.fillna(0).values.reshape(-1, 1),
            df[target_numeric])[0]
    except:
        mi_numeric = np.nan
    
    # Mutual Information with categorical target
    try:
        mi_categorical = mutual_info_classif(col.fillna(0).values.reshape(-1, 1),
            df[target_categorical])[0]
    except:
        mi_categorical = np.nan
    
    # Add to results
    results.append({'Feature': feature,
        '% Missing': round(pct_missing, 2),
        '% Zero': round(pct_zero, 2) if not np.isnan(pct_zero) else 'N/A',
        '% Non-Zero': round(pct_non_zero, 2) if not np.isnan(pct_non_zero) else 'N/A',
        '# Unique': unique_count,
        'Variance': round(variance, 4),
        'Pearson Corr w/ TP': round(corr_pearson, 4),
        'Spearman Corr w/ TP': round(corr_spearman, 4),
        'MI w/ TP': round(mi_numeric, 4),
        'MI w/ UR_Category': round(mi_categorical, 4)})


# Create results DataFrame & save
profile_df = pd.DataFrame(results).sort_values(by='MI w/ TP', ascending=False)
print(profile_df)

profile_df.to_csv('Sparse_Features_Profile.csv', index=False)
```

## Phase 2.1. **Ballast Detection & Reduction**

### **Basic info**

```python
import pandas as pd

df = pd.read_csv("Sparse_Features_Subset.csv")

# Show all column names
print("\n Column Names:")
print(df.columns.tolist())

# Show first 5 rows
print("\n First 5 Rows:")
print(df.head())

# Show basic info
print("\nℹ Dataset Info:")
print(df.info())

# Check missing values
print("\n Missing Values per Column:")
print(df.isnull().sum())
```

#### 1️⃣ **Load & pre-process**: Loads sparse dataset, drops the target column for X, keeps y.
#### 2️⃣ **VarianceThreshold**: Removes features with near-zero variance.
#### 3️⃣ **Correlation filter**: Drops redundant features that are highly correlated (>0.95).
#### 4️⃣ **Mutual Information**: Drops features with MI near zero.
#### 5️⃣ **Entropy filter**: Drops features with very low entropy (no diversity).
#### 6️⃣ **Lasso**: Uses L1 regularization to shrink coefficients to zero for irrelevant features.
#### 7️⃣ **SHAP**: Uses model-aware SHAP values to drop features that have negligible impact.
#### 8️⃣ **PCA**: Final compression step to combine redundancy across features.

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold, mutual_info_classif
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestClassifier
import shap

df = pd.read_csv("Sparse_Features_Subset.csv")

# Use actual target
target_col = 'UR_Category'

# Exclude non-feature columns like 'GUID'
non_feature_cols = ['GUID', target_col]
X = df.drop(columns=non_feature_cols)
y = df[target_col]

print(f"    Initial shape: {X.shape}")

# ---------------------------------------
# 2) Variance Threshold
# ---------------------------------------
print("\n Step 1: Variance Threshold")
var_thresh = VarianceThreshold(threshold=0.01)
X_var = var_thresh.fit_transform(X)
selected_var = X.columns[var_thresh.get_support()]
print(f"Remaining after VarianceThreshold: {len(selected_var)}")
X = X[selected_var]

# ---------------------------------------
# 3) Correlation Filter
# ---------------------------------------
print("\n Step 2: Correlation Filter")
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.95)]
print(f"Highly correlated features dropped: {len(to_drop)}")
X = X.drop(columns=to_drop)

# ---------------------------------------
# 4) Mutual Information
# ---------------------------------------
print("\n Step 3: Mutual Information")
mi = mutual_info_classif(X, y, random_state=42)
mi_series = pd.Series(mi, index=X.columns).sort_values(ascending=False)
mi_selected = mi_series[mi_series > 0.01].index
print(f"Remaining after Mutual Information: {len(mi_selected)}")
X = X[mi_selected]

# ---------------------------------------
# 5) Entropy Filter
# ---------------------------------------
print("\n Step 4: Entropy Filter")
def entropy(series):
    probs = series.value_counts(normalize=True)
    return -(probs * np.log2(probs + 1e-9)).sum()

entropy_scores = X.apply(entropy)
entropy_selected = entropy_scores[entropy_scores > 0.1].index
print(f"Remaining after Entropy filter: {len(entropy_selected)}")
X = X[entropy_selected]

# ---------------------------------------
# 6) Lasso (L1) Feature Selection
# ---------------------------------------
print("\n Step 5: LassoCV")
lasso = LassoCV(cv=5, random_state=42)
lasso.fit(X, y)
coef = pd.Series(lasso.coef_, index=X.columns)
lasso_selected = coef[coef != 0].index
print(f"Remaining after Lasso: {len(lasso_selected)}")
X = X[lasso_selected]


print("\n Step 6: SHAP (Random Forest)")

# Fit RF
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X, y)

# Small sample
X_sample = X.sample(min(500, X.shape[0]), random_state=42)

explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_sample)

# Aggregation for multi-class or binary
if isinstance(shap_values, list):
    per_class = [np.abs(sv).mean(axis=0) for sv in shap_values]
    per_class = np.vstack(per_class)
    shap_importance = per_class.mean(axis=0)
else:
    # Single ndarray (n_samples, n_features, n_classes)
    shap_importance = np.abs(shap_values).mean(axis=0).mean(axis=-1)

print(f"SHAP importance final shape: {shap_importance.shape}")  # Should be (n_features, ...)


shap_series = pd.Series(shap_importance, index=X.columns)
shap_threshold = shap_series.mean()
shap_selected = shap_series[shap_series > shap_threshold].index

print(f"Remaining after SHAP: {len(shap_selected)}")
X = X[shap_selected]

shap.summary_plot(shap_values, X_sample, plot_type="bar")



# ---------------------------------------
# 8) Final feature list
# ---------------------------------------
print(f"\n  Final feature shape: {X.shape}")
print(f" Final features: {list(X.columns)}")

# Save or export reduced feature set if needed
X_final = pd.concat([df['GUID'], X, y], axis=1)
X_final.to_csv("Reduced_Features.csv", index=False)
print("\n   Reduced dataset saved as 'Reduced_Features.csv'")
```

## Complementary PCA + Model Evaluation 

#### PCA for optional dimensionality reduction
#### trains a model (e.g., RandomForestClassifier)
#### evaluates performance (AUC, accuracy, F1, etc.)
#### simple visualization of explained variance & performance

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, recall_score, precision_score
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------
#   Load dataset
# ------------------------------
df = pd.read_csv('Sparse_Features_Subset.csv')

# print("   Columns:", df.columns.tolist())

# Target is 'UR_Category'; GUID is ID
y = df['UR_Category']
X = df.drop(['GUID', 'UR_Category'], axis=1)

print(f"  X shape: {X.shape} | y distribution:\n{y.value_counts()}")

# ------------------------------
#   Train/test split
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print(f"\n    Train shape: {X_train.shape} | Test shape: {X_test.shape}")

# ------------------------------
#   Standardize (needed for PCA)
# ------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ------------------------------
#   Baseline Random Forest
# ------------------------------
print("\n    Baseline Random Forest")
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_test_scaled)
y_prob = rf.predict_proba(X_test_scaled)

# Compute AUC for binary or multi-class
if y_prob.shape[1] > 2:
    auc_baseline = roc_auc_score(y_test, y_prob, multi_class='ovr', average='weighted')
else:
    auc_baseline = roc_auc_score(y_test, y_prob[:, 1])

baseline_results = {"AUC": auc_baseline,
    "Accuracy": accuracy_score(y_test, y_pred),
    "F1": f1_score(y_test, y_pred, average='weighted'),
    "Recall": recall_score(y_test, y_pred, average='weighted'),
    "Precision": precision_score(y_test, y_pred, average='weighted'),
    "Train_Features": X_train.shape[1]}
print(f"\n   Baseline Results:\n{baseline_results}")

# ------------------------------
#    PCA
# ------------------------------
print("\n    Running PCA (retain 95% variance)")
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

print(f" PCA reduced features: {X_train_pca.shape[1]}")

# ------------------------------
#    Visualize Explained Variance
# ------------------------------
plt.figure(figsize=(8, 4))
sns.lineplot(x=range(1, len(pca.explained_variance_ratio_) + 1), y=np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of PCA Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% variance')
plt.legend()
plt.tight_layout()
plt.show()

# ------------------------------
#    Random Forest on PCA data
# ------------------------------
print("\n    Random Forest with PCA")
rf_pca = RandomForestClassifier(n_estimators=100, random_state=42)
rf_pca.fit(X_train_pca, y_train)
y_pred_pca = rf_pca.predict(X_test_pca)
y_prob_pca = rf_pca.predict_proba(X_test_pca)

# Compute AUC for binary or multi-class
if y_prob_pca.shape[1] > 2:
    auc_pca = roc_auc_score(y_test, y_prob_pca, multi_class='ovr', average='weighted')
else:
    auc_pca = roc_auc_score(y_test, y_prob_pca[:, 1])

pca_results = {"AUC": auc_pca,
    "Accuracy": accuracy_score(y_test, y_pred_pca),
    "F1": f1_score(y_test, y_pred_pca, average='weighted'),
    "Recall": recall_score(y_test, y_pred_pca, average='weighted'),
    "Precision": precision_score(y_test, y_pred_pca, average='weighted'),
    "Train_Features": X_train_pca.shape[1]}
print(f"\n     PCA Results:\n{pca_results}")

# ------------------------------
#     Compare & visualize
# ------------------------------
comparison_df = pd.DataFrame([baseline_results, pca_results],
                             index=['Baseline', 'PCA'])

print("\n     Comparison:")
print(comparison_df)

# Barplot
comparison_df[['Accuracy', 'F1', 'Recall', 'Precision']].plot(kind='bar', figsize=(10, 6))
plt.title('Performance Comparison: Baseline vs PCA')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Phase 2.2.1. **Ballast Detection and Pruning (Entropy + MI)**

#### ✅ Entropy & MI calculation for each feature, normalized.
#### ✅ Looping over ballast thresholds (0%–95% in steps of 5%) to prune.
#### ✅ CatBoost for regression (T5_2_TP) & classification (UR_Category).
#### ✅ Trade-off comparison before/after pruning: AUC, Accuracy, F1, Recall, Precision.
#### ✅ Scatter plot: Normalized Entropy vs Normalized MI, color-coded Keep/Drop.

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif
from sklearn.metrics import mean_squared_error, accuracy_score, f1_score, recall_score, precision_score, roc_auc_score
from sklearn.model_selection import train_test_split
from scipy.stats import entropy
from catboost import CatBoostRegressor, CatBoostClassifier
import matplotlib.pyplot as plt

# -----------------------------------
#  Load data
# -----------------------------------
df = pd.read_csv('Sparse_Features_Subset.csv')
df = df.drop(columns=['GUID'])

# Define targets
target_reg = 'T5_2_TP'
target_cls = 'UR_Category'

X = df.drop(columns=[target_reg, target_cls])
y_reg = df[target_reg]
y_cls = df[target_cls]

# -----------------------------------
#  Calculate Entropy & MI
# -----------------------------------
entropy_scores = []
for col in X.columns:
    counts = X[col].value_counts(normalize=True)
    ent = entropy(counts, base=2)
    entropy_scores.append(ent)

# MI for regression & classification
mi_reg = mutual_info_regression(X, y_reg)
mi_cls = mutual_info_classif(X, y_cls)
mi_scores = np.maximum(mi_reg, mi_cls)

# Normalize
norm_entropy = (entropy_scores - np.min(entropy_scores)) / (np.max(entropy_scores) - np.min(entropy_scores))
norm_mi = (mi_scores - np.min(mi_scores)) / (np.max(mi_scores) - np.min(mi_scores))

#  PRODUCT-BASED Ballast Score:
ballast_scores = (1 - norm_entropy) * (1 - norm_mi)

feature_scores = pd.DataFrame({'Feature': X.columns, 'Norm_Entropy': norm_entropy, 'Norm_MI': norm_mi,
    'Ballast_Score': ballast_scores}).sort_values('Ballast_Score', ascending=False)

print("\n   Top ballast features:\n", feature_scores.head(10))

# -----------------------------------
#  Train-test split
# -----------------------------------
X_train, X_test, y_train_reg, y_test_reg, y_train_cls, y_test_cls = train_test_split(
    X, y_reg, y_cls, test_size=0.2, random_state=42)

# -----------------------------------
#  Define CatBoost models
# -----------------------------------

# Add class weights to handle imbalance
# Calculate weights inversely proportional to class frequency
cls_weights = y_train_cls.value_counts(normalize=True)
class_weights = [1 / freq for freq in cls_weights.sort_index()]

regressor = CatBoostRegressor(verbose=0)
classifier = CatBoostClassifier(verbose=0, class_weights=class_weights)

# -----------------------------------
#  Evaluation function
# -----------------------------------
def evaluate_models(X_tr, X_te):
    regressor.fit(X_tr, y_train_reg)
    reg_preds = regressor.predict(X_te)
    mse = mean_squared_error(y_test_reg, reg_preds)

    classifier.fit(X_tr, y_train_cls)
    cls_preds = classifier.predict(X_te)
    cls_probs = classifier.predict_proba(X_te)

    acc = accuracy_score(y_test_cls, cls_preds)
    f1 = f1_score(y_test_cls, cls_preds, average='macro')
    recall = recall_score(y_test_cls, cls_preds, average='macro')
    precision = precision_score(y_test_cls, cls_preds, average='macro')

    try:
        auc = roc_auc_score(y_test_cls, cls_probs, multi_class='ovo', average='macro')
    except:
        auc = np.nan

    return {'Reg_MSE': mse, 'Cls_ACC': acc, 'Cls_F1': f1, 'Cls_Recall': recall,
            'Cls_Precision': precision, 'Cls_AUC': auc}

# -----------------------------------
#  Loop thresholds
# -----------------------------------
results = []

for thresh in np.arange(0.0, 1.0, 0.05):
    quantile = np.quantile(feature_scores['Ballast_Score'], thresh)
    keep_features = feature_scores[feature_scores['Ballast_Score'] <= quantile]['Feature']

    X_tr = X_train[keep_features]
    X_te = X_test[keep_features]

    metrics = evaluate_models(X_tr, X_te)
    metrics['Threshold'] = thresh
    metrics['Num_Features'] = len(keep_features)

    results.append(metrics)

results_df = pd.DataFrame(results)
print("\n   Results for thresholds:\n", results_df)

# -----------------------------------
#   Ballast Map (Product version)
# -----------------------------------
plt.figure(figsize=(10, 6))
# In product version: high (1-NormEntropy)*(1-NormMI) = likely ballast → low NormEntropy & NormMI = red
colors = ['red' if b > np.quantile(ballast_scores, 0.85) else 'green' for b in ballast_scores]

plt.scatter(norm_entropy, norm_mi, c=colors, alpha=0.7)
for i, f in enumerate(X.columns):
    plt.text(norm_entropy[i]+0.005, norm_mi[i]+0.005, f, fontsize=7)

plt.xlabel('Normalized Entropy')
plt.ylabel('Normalized MI')
plt.title('Ballast Map (Product-based)\nRed = Likely Ballast | Green = Likely Informative')
plt.grid(True)
plt.show()


#   Ballast Map (without poin labels)
# -----------------------------------
plt.figure(figsize=(10, 6))
# In product version: high (1-NormEntropy)*(1-NormMI) = likely ballast → low NormEntropy & NormMI = red
colors = ['red' if b > np.quantile(ballast_scores, 0.85) else 'green' for b in ballast_scores]

plt.scatter(norm_entropy, norm_mi, c=colors, alpha=0.7)
# for i, f in enumerate(X.columns):
#     plt.text(norm_entropy[i]+0.005, norm_mi[i]+0.005, f, fontsize=7)

plt.xlabel('Normalized Entropy')
plt.ylabel('Normalized MI')
plt.title('Ballast Map (Product-based)\nRed = Likely Ballast | Green = Likely Informative')
plt.grid(True)
plt.show()
# -----------------------------------
#  Trade-off plot
# -----------------------------------
plt.figure(figsize=(10, 6))
plt.plot(results_df['Threshold']*100, results_df['Cls_ACC'], label='Accuracy')
plt.plot(results_df['Threshold']*100, results_df['Cls_F1'], label='F1 Score')
plt.plot(results_df['Threshold']*100, results_df['Cls_Recall'], label='Recall')
plt.plot(results_df['Threshold']*100, results_df['Cls_Precision'], label='Precision')
plt.xlabel('Ballast Pruning Threshold (%)')
plt.ylabel('Score')
plt.title('Trade-off: Classification Performance vs Ballast Pruning (Product-based)')
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------------
#  Recommended final pruning
# -----------------------------------
# Apply final pruning in the sweet spot (e.g. 0.8–0.9)
final_thresh = 0.86
quantile = np.quantile(feature_scores['Ballast_Score'], final_thresh)
final_features = feature_scores[feature_scores['Ballast_Score'] <= quantile]['Feature']
print(f"\n   Final selected features: {len(final_features)}")
print(final_features.tolist())
```

##  Phase 2.2.2. **PCA Ballast Pruning + Loop + Visual**

#### The script benchmarks how reducing dimensions with PCA (by controlling retained variance) affects Random Forest performance

```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, recall_score, precision_score
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------
#   Load dataset
# ------------------------------
df = pd.read_csv('Sparse_Features_Subset.csv')

y = df['UR_Category']
X = df.drop(['GUID', 'UR_Category'], axis=1)

print(f"  X shape: {X.shape} | y distribution:\n{y.value_counts()}")

# ------------------------------
#   Train/test split
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(f"\n    Train shape: {X_train.shape} | Test shape: {X_test.shape}")

# ------------------------------
#   Standardize (needed for PCA)
# ------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ------------------------------
#   Baseline Random Forest
# ------------------------------
print("\n    Baseline Random Forest")
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_test_scaled)
y_prob = rf.predict_proba(X_test_scaled)

if y_prob.shape[1] > 2:
    auc_baseline = roc_auc_score(y_test, y_prob, multi_class='ovr', average='weighted')
else:
    auc_baseline = roc_auc_score(y_test, y_prob[:, 1])

baseline_results = {
    "AUC": auc_baseline,
    "Accuracy": accuracy_score(y_test, y_pred),
    "F1": f1_score(y_test, y_pred, average='weighted'),
    "Recall": recall_score(y_test, y_pred, average='weighted'),
    "Precision": precision_score(y_test, y_pred, average='weighted'),
    "Num_Components": X_train.shape[1],
    "Variance_Retained": 1.0
}
print(f"\n   Baseline Results:\n{baseline_results}")

# ------------------------------
#   Loop over PCA variance thresholds
# ------------------------------
print("\n   PCA Threshold Loop")
results = []
for thresh in np.arange(0.05, 0.95, 0.05):
    pca = PCA(n_components=thresh, random_state=42)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    rf_pca = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_pca.fit(X_train_pca, y_train)
    y_pred_pca = rf_pca.predict(X_test_pca)
    y_prob_pca = rf_pca.predict_proba(X_test_pca)

    if y_prob_pca.shape[1] > 2:
        auc_pca = roc_auc_score(y_test, y_prob_pca, multi_class='ovr', average='weighted')
    else:
        auc_pca = roc_auc_score(y_test, y_prob_pca[:, 1])

    results.append({
        'AUC': auc_pca,
        'Accuracy': accuracy_score(y_test, y_pred_pca),
        'F1': f1_score(y_test, y_pred_pca, average='weighted'),
        'Recall': recall_score(y_test, y_pred_pca, average='weighted'),
        'Precision': precision_score(y_test, y_pred_pca, average='weighted'),
        'Num_Components': X_train_pca.shape[1],
        'Variance_Retained': thresh
    })

# ------------------------------
#   Visualize Results
# ------------------------------
results_df = pd.DataFrame(results)
print("\n    PCA Pruning Results:\n", results_df)

plt.figure(figsize=(10, 6))
sns.lineplot(x=results_df['Variance_Retained'], y=results_df['AUC'], marker='o', label='AUC')
sns.lineplot(x=results_df['Variance_Retained'], y=results_df['Accuracy'], marker='o', label='Accuracy')
plt.axhline(y=baseline_results['AUC'], color='red', linestyle='--', label='Baseline AUC')
plt.xlabel('Variance Retained (%)')
plt.ylabel('Score')
plt.title('Performance vs PCA Variance Retained')
plt.legend()
plt.tight_layout()
plt.show()

# Optional: print best performing
best_row = results_df.loc[results_df['AUC'].idxmax()]
print(f"\n Best PCA config:\n{best_row}")
```

## Phase 2.2.3. **VarianceThreshold & Correlation Filter** - Loop, Train, Compare, Plot

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (roc_auc_score, accuracy_score, f1_score, recall_score, precision_score)
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Sparse_Features_Subset.csv")

target_col = "UR_Category"
non_feature_cols = ["GUID", target_col]

X_orig = df.drop(columns=non_feature_cols)
y = df[target_col]

print(f"Initial shape: {X_orig.shape}")

# ===========================
# Split once for consistency
# ===========================
X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_orig, y, test_size=0.2, random_state=42, stratify=y)

# ===========================
# 1) VarianceThreshold loop
# ===========================
var_thresholds = np.arange(0.0, 0.45, 0.02)
var_results = []

print("\n    VarianceThreshold loop")
for thresh in var_thresholds:
    selector = VarianceThreshold(threshold=thresh)
    X_train = selector.fit_transform(X_train_full)
    X_test = selector.transform(X_test_full)

    num_features = X_train.shape[1]

    if num_features == 0:
        continue  # skip empty

    model = RandomForestClassifier(random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    if y_proba.shape[1] > 2:
        auc = roc_auc_score(y_test, y_proba, multi_class="ovr")
    else:
        auc = roc_auc_score(y_test, y_proba[:, 1])

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")

    var_results.append({"Threshold": thresh,
        "Num_Features": num_features,
        "AUC": auc,
        "Accuracy": acc,
        "F1": f1})

var_df = pd.DataFrame(var_results)
print("\nVarianceThreshold results:")
print(var_df)

# ===========================
# Plot VarianceThreshold
# ===========================
plt.figure(figsize=(12, 6))
sns.lineplot(data=var_df, x="Threshold", y="AUC", marker="o", label="AUC")
sns.lineplot(data=var_df, x="Threshold", y="Accuracy", marker="s", label="Accuracy")
sns.lineplot(data=var_df, x="Threshold", y="F1", marker="^", label="F1")
plt.title("VarianceThreshold: Metrics vs. Threshold")
plt.xlabel("Variance Threshold")
plt.ylabel("Metric Score")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.lineplot(data=var_df, x="Threshold", y="Num_Features", marker="o")
plt.title("VarianceThreshold: Features Remaining vs. Threshold")
plt.xlabel("Variance Threshold")
plt.ylabel("Number of Features")
plt.show()

# ===========================
# 2) Correlation Filter loop
# ===========================
corr_thresholds = np.arange(0.30, 0.50, 0.02)
corr_results = []

print("\n    Correlation Filter loop")
for thresh in corr_thresholds:
    corr_matrix = X_train_full.corr().abs()
    upper = corr_matrix.where(
        np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)    )

    to_drop = [column for column in upper.columns if any(upper[column] > thresh)]

    X_train = X_train_full.drop(columns=to_drop)
    X_test = X_test_full.drop(columns=to_drop)

    num_features = X_train.shape[1]

    if num_features == 0:
        continue

    model = RandomForestClassifier(random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    # Safe AUC for binary or multiclass
    if y_proba.shape[1] > 2:
        auc = roc_auc_score(y_test, y_proba, multi_class="ovr")
    else:
        auc = roc_auc_score(y_test, y_proba[:, 1])

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")

    corr_results.append({"Threshold": thresh, "Num_Features": num_features,
        "AUC": auc, "Accuracy": acc, "F1": f1})

corr_df = pd.DataFrame(corr_results)
print("\nCorrelation Filter results:")
print(corr_df)

# ===========================
#   Plot Correlation Filter
# ===========================
plt.figure(figsize=(12, 6))
sns.lineplot(data=corr_df, x="Threshold", y="AUC", marker="o", label="AUC")
sns.lineplot(data=corr_df, x="Threshold", y="Accuracy", marker="s", label="Accuracy")
sns.lineplot(data=corr_df, x="Threshold", y="F1", marker="^", label="F1")
plt.title("Correlation Filter: Metrics vs. Threshold")
plt.xlabel("Correlation Threshold")
plt.ylabel("Metric Score")
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.lineplot(data=corr_df, x="Threshold", y="Num_Features", marker="o")
plt.title("Correlation Filter: Features Remaining vs. Threshold")
plt.xlabel("Correlation Threshold")
plt.ylabel("Number of Features")
plt.show()
```

## Phase 2.2.4. **Lasso (L1) Ballast Pruning Code**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split  # ✅ fix for train_test_split
from sklearn.linear_model import LassoCV, LogisticRegressionCV
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score

df = pd.read_csv("Sparse_Features_Subset.csv")

target_col = "UR_Category"
non_feature_cols = ["GUID", target_col]

X_orig = df.drop(columns=non_feature_cols)
y = df[target_col]

print(f"    Initial shape: {X_orig.shape}")

# ================================
#  Split data once for consistency
# ================================
X_train_full, X_test_full, y_train, y_test = train_test_split(
    X_orig, y, test_size=0.2, random_state=42, stratify=y)

print(f"   Train shape: {X_train_full.shape}")
print(f"   Test shape: {X_test_full.shape}")
# ================================
#  Define alphas to test
# ================================
lasso_alphas = np.logspace(-5, 0, 15)
lasso_results = []

for alpha in lasso_alphas:
    
    # -------------------------------
    # Fit base model
    # -------------------------------
    if len(np.unique(y_train)) == 2:
        # Binary → Logistic Regression with L1
        base_model = LogisticRegressionCV(Cs=[1/alpha], penalty="l1", solver="saga",
            cv=5, random_state=42, n_jobs=-1, max_iter=5000)
    else:
        # Regression → LassoCV
        base_model = LassoCV(alphas=[alpha], cv=5, random_state=42)

        base_model.fit(X_train_full, y_train)

    # -------------------------------
    #   Select non-zero coefficients
    # -------------------------------
    coef = base_model.coef_

    # Handle multi-class: coef_ shape (n_classes, n_features)
    if coef.ndim > 1:
        coef = np.sum(np.abs(coef), axis=0)
    else:
        coef = np.abs(coef)

    # Mask and keep columns with non-zero coefficients
    feature_mask = coef > 1e-8
    selected_cols = X_train_full.columns[feature_mask]

    num_features = len(selected_cols)

    if num_features == 0:
        print(f"    Alpha {alpha:.5f}: No features selected, skipping.")
        continue

        # Slice DataFrames directly to keep feature names
    X_train = X_train_full[selected_cols]
    X_test = X_test_full[selected_cols]
    
    # -------------------------------
    # Train final Random Forest
    # -------------------------------
    model = RandomForestClassifier(random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    # -------------------------------
    # Compute metrics safely
    # -------------------------------
    if len(np.unique(y_test)) > 2:
        auc = roc_auc_score(y_test, y_proba, multi_class="ovr")
    else:
        auc = roc_auc_score(y_test, y_proba[:, 1])

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")

    lasso_results.append({"Alpha": alpha, "Num_Features": num_features,
        "AUC": auc, "Accuracy": acc, "F1": f1})

# ================================
#  Results DataFrame
# ================================
lasso_df = pd.DataFrame(lasso_results)
print("\n     Lasso/Logistic L1 results:")
print(lasso_df)

# ================================
#  Plot performance metrics
# ================================
plt.figure(figsize=(8, 5))
sns.lineplot(data=lasso_df, x="Alpha", y="AUC", marker="o", label="AUC")
sns.lineplot(data=lasso_df, x="Alpha", y="Accuracy", marker="s", label="Accuracy")
sns.lineplot(data=lasso_df, x="Alpha", y="F1", marker="^", label="F1")
plt.xscale("log")
plt.title("Lasso/Logistic L1: Metrics vs. Alpha")
plt.xlabel("Alpha (Regularization Strength)")
plt.ylabel("Metric Score")
plt.legend()
plt.tight_layout()
plt.show()

# ================================
# Plot # features vs alpha
# ================================
plt.figure(figsize=(8, 5))
sns.lineplot(data=lasso_df, x="Alpha", y="Num_Features", marker="o")
plt.xscale("log")
plt.title("Lasso/Logistic L1: Features Remaining vs. Alpha")
plt.xlabel("Alpha (Regularization Strength)")
plt.ylabel("Number of Features")
plt.tight_layout()
plt.show()

# ================================
# Plot “Number of Features vs AUC”
# ================================
plt.figure(figsize=(8, 5))
sns.lineplot(data=lasso_df, x="Num_Features", y="AUC", marker="o", linewidth=2.5)

plt.title("Number of Features vs AUC (Lasso/Logistic L1)", fontsize=16)
plt.xlabel("Number of Features Selected", fontsize=12)
plt.ylabel("AUC Score", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Phase 2.2.4. **LightGBM + SHAP pruning** 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import lightgbm as lgb
import shap
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, recall_score
import matplotlib.pyplot as plt

df = pd.read_csv("Sparse_Features_Subset1.csv")
print("Original shape:", df.shape)

# ===============================
# 2. Define target and ID columns
# ===============================
TARGET = "SA_CHANGE_CODE"
ID_COLS = ["GUID"]  # Optional

# ===============================
# 3. Split data
# ===============================
X = df.drop(ID_COLS + [TARGET], axis=1)
y = df[TARGET]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

# ===============================
# 4. Fit LightGBM model
# ===============================
model = lgb.LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

# ===============================
# 5. Compute SHAP values
# ===============================
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_train)

print("shap_values type:", type(shap_values))

if isinstance(shap_values, list):
    print("Detected: list of arrays")
    print("shap_values[0] shape:", shap_values[0].shape)
    # For list (multi-class): mean each class's abs SHAP, then mean over classes
    mean_shap = np.mean([np.abs(sv).mean(axis=0) for sv in shap_values], axis=0)
elif isinstance(shap_values, np.ndarray) and shap_values.ndim == 3:
    print("Detected: ndarray with 3 dimensions")
    print("shap_values shape:", shap_values.shape)
    # For ndarray (multi-class): mean over samples, then classes
    mean_shap = np.abs(shap_values).mean(axis=0).mean(axis=1)
else:
    print("Detected: ndarray with shape", shap_values.shape)
    mean_shap = np.abs(shap_values).mean(axis=0)

print("mean_shap shape:", mean_shap.shape)

# ===============================
# 6. Rank features by importance
# ===============================
shap_importance = pd.Series(mean_shap, index=X_train.columns).sort_values(ascending=False)
print("\nTop SHAP features:\n", shap_importance.head(10))

# ===============================
# 7. Sweep thresholds & evaluate
# ===============================
thresholds = np.arange(0.0, 0.15, 0.005)
results = []

for threshold in thresholds:
    keep_features = shap_importance[shap_importance >= threshold].index.tolist()
    
    if len(keep_features) == 0:
        print(f"Skipping threshold {threshold:.2f}: no features remain.")
        continue

    X_train_subset = X_train[keep_features]
    X_test_subset = X_test[keep_features]

    model_subset = lgb.LGBMClassifier(random_state=42)
    model_subset.fit(X_train_subset, y_train)
 
    y_pred_proba = model_subset.predict_proba(X_test_subset)
    y_pred = model_subset.predict(X_test_subset)

    # AUC
    if y_pred_proba.shape[1] == 2:
        auc = roc_auc_score(y_test, y_pred_proba[:, 1])
    else:
        auc = roc_auc_score(y_test, y_pred_proba, multi_class="ovr")

    # Accuracy, F1, Recall
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")
    rec = recall_score(y_test, y_pred, average="weighted")

    results.append({"threshold": threshold, "n_features": len(keep_features),
        "auc": auc, "accuracy": acc, "f1": f1, "recall": rec})

# ===============================
# 8. Show results
# ===============================
results_df = pd.DataFrame(results)
print("\n Summary of Results:\n", results_df)


# Plot AUC vs Threshold
plt.figure(figsize=(10,5))
plt.plot(results_df['threshold'], results_df['auc'], marker='o')
plt.title('AUC vs SHAP Threshold')
plt.xlabel('SHAP Threshold')
plt.ylabel('AUC')
plt.grid(True)
plt.show()

# Plot Number of Features vs Threshold
plt.figure(figsize=(10,5))
plt.plot(results_df['threshold'], results_df['n_features'], marker='o', color='red')
plt.title('Number of Features vs SHAP Threshold')
plt.xlabel('SHAP Threshold')
plt.ylabel('Number of Features')
plt.grid(True)
plt.show()

# Multi-metric plots
plt.figure(figsize=(10, 5))
plt.plot(results_df['threshold'], results_df['auc'], marker='o', label='AUC', color='blue')
plt.plot(results_df['threshold'], results_df['accuracy'], marker='o', label='Accuracy', color='green')
plt.plot(results_df['threshold'], results_df['f1'], marker='o', label='F1 Score', color='orange')
plt.plot(results_df['threshold'], results_df['recall'], marker='o', label='Recall', color='red')
plt.title('Model Metrics vs SHAP Threshold')
plt.xlabel('SHAP Threshold')
plt.ylabel('Metric Value')
plt.ylim(0.60, 0.85)  
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

## Phase 2.2.5. **FULL Information Bottleneck Ballast Pruning**

#### ✔️ Uses mutual info as IB proxy: *I(X;Y)* = how much a feature explains about target → high = useful, low = ballast.
#### ✔️ Ballast score = 1 − Norm_MI.
#### ✔️ Loops over thresholds (0.05–0.95) → prunes features with high ballast.
#### ✔️ Evaluates both regression (MSE) and classification (Acc, F1, Recall, AUC).
#### ✔️ Includes ballast map & trade-off curve for interpretability.



```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, f1_score, recall_score, precision_score, roc_auc_score
from catboost import CatBoostRegressor, CatBoostClassifier
import matplotlib.pyplot as plt

# ------------------------------
# LOAD DATA
# ------------------------------
df = pd.read_csv('Sparse_Features_Subset.csv')
df = df.drop(columns=['GUID'])  # Keep only relevant

# Define targets
target_reg = 'T5_2_TP'
target_cls = 'UR_Category'

X = df.drop(columns=[target_reg, target_cls])
y_reg = df[target_reg]
y_cls = df[target_cls]

# ------------------------------
# CALCULATE Mutual Info for IB
# ------------------------------

# For IB: I(X;Y) ~ MI (informative). Low MI → likely ballast under IB.
mi_reg = mutual_info_regression(X, y_reg)
mi_cls = mutual_info_classif(X, y_cls)
mi_ib = np.maximum(mi_reg, mi_cls)

# Normalize MI scores
norm_mi_ib = (mi_ib - np.min(mi_ib)) / (np.max(mi_ib) - np.min(mi_ib))

# IB ballast score: inverse of MI
ballast_scores_ib = 1 - norm_mi_ib

feature_scores_ib = pd.DataFrame({
    'Feature': X.columns,
    'Norm_MI_IB': norm_mi_ib,
    'Ballast_Score_IB': ballast_scores_ib
}).sort_values('Ballast_Score_IB', ascending=False)

print("\n   Top ballast features (IB):\n", feature_scores_ib.head(10))

# ------------------------------
# TRAIN-TEST SPLIT
# ------------------------------
X_train, X_test, y_train_reg, y_test_reg, y_train_cls, y_test_cls = train_test_split(
    X, y_reg, y_cls, test_size=0.2, random_state=42)

# ------------------------------
# DEFINE CatBoost MODELS
# ------------------------------
# Handle class imbalance
cls_weights = y_train_cls.value_counts(normalize=True)
class_weights = [1 / freq for freq in cls_weights.sort_index()]

regressor = CatBoostRegressor(verbose=0)
classifier = CatBoostClassifier(verbose=0, class_weights=class_weights)

# ------------------------------
# EVALUATION FUNCTION
# ------------------------------
def evaluate_models(X_tr, X_te):
    regressor.fit(X_tr, y_train_reg)
    reg_preds = regressor.predict(X_te)
    mse = mean_squared_error(y_test_reg, reg_preds)

    classifier.fit(X_tr, y_train_cls)
    cls_preds = classifier.predict(X_te)
    cls_probs = classifier.predict_proba(X_te)

    acc = accuracy_score(y_test_cls, cls_preds)
    f1 = f1_score(y_test_cls, cls_preds, average='macro')
    recall = recall_score(y_test_cls, cls_preds, average='macro')
    precision = precision_score(y_test_cls, cls_preds, average='macro')

    try:
        auc = roc_auc_score(y_test_cls, cls_probs, multi_class='ovo', average='macro')
    except:
        auc = np.nan

    return {'Reg_MSE': mse, 'Cls_ACC': acc, 'Cls_F1': f1, 'Cls_Recall': recall,
            'Cls_Precision': precision, 'Cls_AUC': auc}

# ------------------------------
# LOOP THRESHOLDS
# ------------------------------
results_ib = []

for thresh in np.arange(0.0, 1.0, 0.05):
    quantile = np.quantile(feature_scores_ib['Ballast_Score_IB'], thresh)
    keep_features = feature_scores_ib[feature_scores_ib['Ballast_Score_IB'] <= quantile]['Feature']

    X_tr = X_train[keep_features]
    X_te = X_test[keep_features]

    metrics = evaluate_models(X_tr, X_te)
    metrics['Threshold'] = thresh
    metrics['Num_Features'] = len(keep_features)

    results_ib.append(metrics)

results_df_ib = pd.DataFrame(results_ib)
print("\n   Results for thresholds (IB):\n", results_df_ib)

# ------------------------------
# BALLAST MAP for IB
# ------------------------------
plt.figure(figsize=(10, 6))
colors = ['red' if s > np.quantile(ballast_scores_ib, 0.85) else 'green' for s in ballast_scores_ib]
plt.scatter(norm_mi_ib, ballast_scores_ib, c=colors, alpha=0.7)

for i, f in enumerate(X.columns):
    plt.text(norm_mi_ib[i]+0.005, ballast_scores_ib[i]+0.005, f, fontsize=7)

plt.xlabel('Normalized MI (IB)')
plt.ylabel('IB Ballast Score (1-Norm_MI)')
plt.title('Ballast Map (Information Bottleneck)\nRed = Likely Ballast | Green = Likely Informative')
plt.grid(True)
plt.show()

# ------------------------------
# TRADE-OFF PLOT
# ------------------------------
plt.figure(figsize=(10, 6))
plt.plot(results_df_ib['Threshold']*100, results_df_ib['Cls_ACC'], label='Accuracy')
plt.plot(results_df_ib['Threshold']*100, results_df_ib['Cls_F1'], label='F1 Score')
plt.plot(results_df_ib['Threshold']*100, results_df_ib['Cls_Recall'], label='Recall')
plt.plot(results_df_ib['Threshold']*100, results_df_ib['Cls_Precision'], label='Precision')
plt.xlabel('Ballast Pruning Threshold (%) (IB)')
plt.ylabel('Score')
plt.title('Trade-off: Classification Performance vs Ballast Pruning (Information Bottleneck)')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------
# FINAL SELECTED FEATURES EXAMPLE
# ------------------------------
final_thresh = 0.85
quantile = np.quantile(feature_scores_ib['Ballast_Score_IB'], final_thresh)
final_features = feature_scores_ib[feature_scores_ib['Ballast_Score_IB'] <= quantile]['Feature']

print(f"\n   Final selected features (IB): {len(final_features)}")
print(final_features.tolist())
```
